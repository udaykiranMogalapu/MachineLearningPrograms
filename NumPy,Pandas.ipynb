```python
import numpy as np
a1=np.array([[1,2,3],[4,5,6]])
print(a1[0,1])
print(a1)
print(a1.shape)
a2=np.array([[4,5,6],[7,8,9]])
print(a1+a2)
print(a2[-1])
sq=np.sqrt(a1)
print(sq)
s=a1.sum()
mi=a1.min()
ma=a2.max()
me=a2.mean()
print(s,mi,ma,me)
ax=np.sum(a1,axis=0)
ax2=np.mean(a2,axis=1)
print(ax)
print(ax2)
a3=np.array([[1,2,3,4],[1,3,2,5],[4,3,2,1]])
m=np.matmul(a1,a3)
print(m)
a4=np.array([[1,2],[3,4]])
ainv=np.linalg.inv(a4)
print(ainv)
```
